```python
# imdb_df['belongs_to_collection'][0]
# imdb_df['genres'][0]
#checking if the numerical data is good or not.
imdb_df.describe()
  #to be able to reverse changes that are not required, we can create a copy of the dataframe on which we can do our operations.
df = imdb_df.copy()
```

```python
#There are records with runtime = 0 and records with budget = 0, revenue = 1.
#So, in the budget column, outlier values other than 0 are present as well and all those values should be replaced with the median value too!
# imdb_df[imdb_df['runtime'] == 0]
imdb_df[(imdb_df['budget'] > 0) & (imdb_df['budget'] < 1000)].head(30)
# imdb_df[imdb_df['revenue'] == 1]
```

```python
#replacing the outlier data with median in RUNTIME column.
#a new variation of df.loc. Instead of selecting all the rows, we select rows in the basis of a condition and within those set of rows we want a specific column.
runtime_new = df.loc[df['runtime'] > 0, 'runtime'].median()
#I want to find the median of all the data with runtime > 0 because ==0 has no median.
runtime_new #the value of runtime_new = 104 which is the median value. Now I will apply this value to all the data with runtime value = 0
df.loc[df['runtime'] == 0, 'runtime'] = runtime_new
```

```python
#replacing the outlier data with median in BUDGET column.
median_budget = df.loc[df['budget']>0, 'budget'].median()
df.loc[df['budget'] == 0, 'budget'] = median_budget
```
